```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen import Lattice 
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.structure import Structure, Molecule
from pymatgen.symmetry.structure import SymmetrizedStructure
from pymatgen.core.structure import PeriodicSite
from pymatgen.core.operations import SymmOp
from pymatgen.util.coord import find_in_coord_list, pbc_diff
from pprint import pprint

with MPRester("QpfJCrDNtAVubMxG") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-7478")
    
    # Creation of the CIF file
    w = CifWriter(structure)
    CIF=w.write_file('RbTmO2_mp-7478.cif')
#Réseau cristallin    
print("Material :RbTmO2")
print("ID :mp-7478")
print("")
SgA=SpacegroupAnalyzer(structure)
print("Groupe spatiale:"+ SgA.get_space_group_symbol())
#print("Hall :"+ SpacegroupAnalyzer(structure).get_hall())
print("Groupe ponctuel: "+ SgA.get_point_group_symbol())
print("Système cristallin: "+ SgA.get_crystal_system())
print("Type de maille: "+ SgA.get_lattice_type())
print("")
print("Vecteurs de base du réseau direct:")
print("__________________________________")
print(" ")
print("Coordonnées des vecteurs de base:")
sl=structure.lattice
print(sl)
print(" ")
print ("Angles entre les vecteurs")
print('\u03B1','=',sl.angles[0],'[°]')
print('\u03B2',"=",sl.angles[1],'[°]')
print('\u03B3',"=",sl.angles[2],'[°]')
print(" ")
print("Longueurs des différents vecteurs")
print("a=",sl.lengths[0],'[Å]')
print("b=",sl.lengths[1],'[Å]')
print("c=",sl.lengths[2],'[Å]')
print(" ")
print("Volume de la maille du réseau direct")
print('\u03A9',"=",sl.volume,'[Å³]')
print(" ")
print("")
print("Vecteurs de base du réseau réciproque:")
print("______________________________________ ")
print("")
print('Coordonnées des vecteurs de base:')
print(sl.reciprocal_lattice)
print(" ")
print("Longueurs des différents vecteurs")
print("a*=",sl.reciprocal_lattice.lengths[0],'1/Å')
print("b*=",sl.reciprocal_lattice.lengths[1],'1/Å')
print("c*=",sl.reciprocal_lattice.lengths[2],'1/Å')
print(" ")
print("Volume de la maille du réseau réciproque")
print('\u03A9*',"=",sl.reciprocal_lattice.volume,'[1/Å³]')
print(" ")
print("")
print("Résumé pour le réseau direct permettant en plus d'avoir les coordonnées des différents atomes via PeriodicSite ")
print("___________________________________________________________________________________________________________")
print("")
pprint(m.get_data("mp-7478",prop="structure"))
    
    
```
