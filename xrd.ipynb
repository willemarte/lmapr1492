```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen import Lattice 
from pymatgen.core.structure import Structure, Molecule
from pymatgen.symmetry.structure import SymmetrizedStructure
from pymatgen.core.structure import PeriodicSite
from pymatgen.core.operations import SymmOp
from pymatgen.util.coord import find_in_coord_list, pbc_diff
from pprint import pprint
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
import os
import json
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("QpfJCrDNtAVubMxG") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-7478")

    
XRD=XRDCalculator('CuKa1') 

# Calculates the diffraction pattern for a structure.
# xrd = DiffractionPattern(x, y, hkls, d_hkls)
XRDPattern=XRD.get_pattern(structure)
# x: Two theta angles
# y: Intensities
# hkls: [{"hkl": (h, k, l), "multiplicity": mult}],
#       where {"hkl": (h, k, l), "multiplicity": mult}
#       is a dict of Miller
#       indices for all diffracted lattice facets contributing to each
#       intensity.
# d_hkls: List of interplanar spacings

print("Dans cette partie nous allons déterminer les 3 premiers pics du diffractogramme pour le cristal de RbTmO2 pour un rayonnement incident Cu Kα  de longueur d’onde 1.5406 Å")
print("")
print("Les différentes valeurs des trois premiers pics du diffractogramme sont reprises dans le tableau ci-dessous:")
print("")
print("Indice hkl","","|","","","","Angle 2\u0398","","","","","","","","","","|","","Intensitée du pic","","","","|","Paramètre dhkl désignant la distance")
print("","","","","","","","","","","","","|","","","","","","","","","","","","","","","","","","","","","","|","","","","","","","","","","","","","","","","","","","","","","","|","interplanaire pour la famille de plans (hkl)")
print("")
print(XRDPattern.hkls[0][0]['hkl'],"","","|","","",XRDPattern.x[0],"°","|","","",XRDPattern.y[0],"%","|",XRDPattern.d_hkls[0])
print(XRDPattern.hkls[1][0]['hkl'],"","","|",XRDPattern.x[1],"°","|","","",XRDPattern.y[1],"%","|",XRDPattern.d_hkls[1])
print(XRDPattern.hkls[2][0]['hkl'],"","","|",XRDPattern.x[2],"°","|","",XRDPattern.y[2],"%","|",XRDPattern.d_hkls[2])
print("")
print("")
print("Ce graphe permet de visualiser la rapport d'intensité entre les 3 premiers pics du diffractogramme et le 5ème pic du diffractogramme qui représente une intensité de 100%"  )
print("_______________________________________________________________________________________________________________________________")
XRD.show_plot(structure)
              
print("")
print("Ce graphe montre les 3 premiers pics du diffractogramme")
print("_______________________________________________________")
XRD.show_plot(structure,two_theta_range=(0,31))
```
